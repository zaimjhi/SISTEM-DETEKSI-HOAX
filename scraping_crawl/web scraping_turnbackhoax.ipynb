```python
import requests
from bs4 import BeautifulSoup
import csv
import time

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
}

def get_article_content(link):
    try:
        res = requests.get(link, headers=headers)
        soup = BeautifulSoup(res.text, 'lxml')
        div_content = soup.find('div', class_='entry-content mh-clearfix')
        paragraphs = div_content.find_all('p')
        content = ' '.join([p.get_text(strip=True) for p in paragraphs])
        return content
    except Exception as e:
        print(f"Error artikel: {link} | {e}")
        return ''

url_first = 'https://turnbackhoax.id/?s=timnas+indonesia'
res_first = requests.get(url_first, headers=headers)
soup_first = BeautifulSoup(res_first.text, 'lxml')

page_numbers = []
for a in soup_first.find_all('a'):
    try:
        num = int(a.text.strip())
        page_numbers.append(num)
    except:
        continue

last_page = max(page_numbers)
print(f"ğŸ” Total halaman ditemukan {last_page}")

with open('turnbackhoax.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Title', 'Link', 'Time', 'Content'])

    for page in range(1, last_page + 1):
        print(f"\nğŸ“„ Scraping halaman {page}...")
        url = f'https://turnbackhoax.id/page/{page}/?s=timnas+indonesia'
        try:
            res = requests.get(url, headers=headers)
            soup = BeautifulSoup(res.text, 'lxml')
            articles = soup.find_all('article')

            for art in articles:
                try:
                    a_tag = art.find('h3', class_='entry-title').find('a')
                    title = a_tag.text.strip()
                    link = a_tag['href']
                    time_info = art.find('span', class_='mh-meta-date').text.strip()
                    content = get_article_content(link)
                    writer.writerow([title, link, time_info, content])
                    print(f"âœ… {title}")
                    time.sleep(1)
                except Exception as e:
                    print(f"Error parsing artikel di halaman {page}: {e}")
                    continue

            time.sleep(1)
        except Exception as e:
            print(f"Gagal akses page {page}: {e}")
            continue
```
